```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("db/task-3.csv")
df
```

```python
df.columns
```

```python

plt.hist(df['SepalLengthCm'], bins=20, edgecolor='black')
plt.title('Distribution of Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Species'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(12, 8))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    plt.hist(df[feature], bins=20, edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
mean_by_species = df.groupby('Species')[features].mean()

mean_by_species.plot(kind='bar', figsize=(10,6), edgecolor='black')
plt.title('Mean Feature Values by Iris Species')
plt.ylabel('Mean Value (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python
df['Species'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
